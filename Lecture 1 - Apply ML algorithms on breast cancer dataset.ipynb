```python
import pandas as pd
import numpy as np
```

### Load dataset using Pandas

```python
dataframe = pd.read_csv("data/METABRIC_RNA_Mutation.csv", sep=",")
```

```python
dataframe
```

```python
# Drop all rows which contain empty cells
dataframe_row_clean = dataframe.dropna(axis=0)
```

```python
# Drop columns which contain empty cells
dataframe_clean = dataframe_row_clean.dropna(axis=1)
```

```python
dataframe_clean
```

```python
# Save clean dataset
dataframe_clean.to_csv("data/METABRIC_RNA_Mutation_cleaned.csv", index=False)
```

### Convert all text values into integers

```python
for col in dataframe_clean.columns.tolist():
    dtype = dataframe_clean[col].dtype
    if dtype == "object":
        dataframe_clean[col] = dataframe_clean[col].astype('category')
```

```python
cat_columns = dataframe_clean.select_dtypes(['category']).columns
```

```python
dataframe_clean[cat_columns] = dataframe_clean[cat_columns].apply(lambda x: x.cat.codes)
```

```python
dataframe_clean.to_csv("data/METABRIC_RNA_Mutation_features.csv", index=False)
```

```python
dataframe_clean
```

### Separate feature columns and target/class

```python
target_col = "overall_survival"
target_overall_survival = dataframe_clean[target_col]
```

```python
target_overall_survival
```

```python
dataframe_clean_features = dataframe_clean.drop([target_col], axis=1)
```

```python
dataframe_clean_features
```

### Apply Scikit learn ML algorithms for classification

```python
from sklearn.model_selection import train_test_split
```

```python
# Split original dataset into training and test parts
X_train, X_test, y_train, y_test = train_test_split(dataframe_clean_features, target_overall_survival, test_size=0.2, random_state=0)
```

### Linear model

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Train model
linear_model = LogisticRegression().fit(X_train, y_train)

# Accuracy on training data
train_lm_clf_score = linear_model.score(X_train, y_train)
print("Train classification accuracy: {}".format(str(np.round(train_lm_clf_score, 2))))

# Accuracy on test data
test_lm_clf_score = linear_model.score(X_test, y_test)
print("Test classification accuracy: {}".format(str(np.round(test_lm_clf_score, 2))))
```

```python
# Linear model's coefficients - W vector
linear_model.coef_.shape
```

```python
# Linear model's intercept
linear_model.intercept_
```

### Nearest neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier().fit(X_train, y_train)

train_knn_clf_score = knn_model.score(X_train, y_train)
print("Train classification accuracy: {}".format(str(np.round(train_knn_clf_score, 2))))

# Accuracy on test data
test_knn_clf_score = knn_model.score(X_test, y_test)
print("Test classification accuracy: {}".format(str(np.round(test_knn_clf_score, 2))))
```

### SVM

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC().fit(X_train, y_train)

train_svc_clf_score = svc_model.score(X_train, y_train)
print("Train classification accuracy: {}".format(str(np.round(train_svc_clf_score, 2))))

# Accuracy on test data
test_svc_clf_score = svc_model.score(X_test, y_test)
print("Test classification accuracy: {}".format(str(np.round(test_svc_clf_score, 2))))
```

### Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier().fit(X_train, y_train)

train_dt_clf_score = dt_model.score(X_train, y_train)
print("Train classification accuracy: {}".format(str(np.round(train_dt_clf_score, 2))))

# Accuracy on test data
test_dt_clf_score = dt_model.score(X_test, y_test)
print("Test classification accuracy: {}".format(str(np.round(test_dt_clf_score, 2))))
```

### Ensemble model

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbt_model = GradientBoostingClassifier().fit(X_train, y_train)

train_gbt_clf_score = gbt_model.score(X_train, y_train)
print("Train classification accuracy: {}".format(str(np.round(train_gbt_clf_score, 2))))

# Accuracy on test data
test_gbt_clf_score = gbt_model.score(X_test, y_test)
print("Test classification accuracy: {}".format(str(np.round(test_gbt_clf_score, 2))))
```

### Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'kernel':('linear', 'poly', 'sigmoid', 'rbf'), 'C':[0.5, 1, 5, 10, 15], 'gamma': ['scale', 'auto']}
```

```python
svc = SVC()
tuned_svc_clf = GridSearchCV(svc, parameters)
tuned_svc_clf.fit(X_train, y_train)
GridSearchCV(estimator=SVC(), param_grid=parameters)
```

```python
tuned_svc_clf.cv_results_
```

```python
# Hyperparameter setting that gives best accuracy
tuned_svc_clf.best_params_
```

```python
train_tuned_svc_score = tuned_svc_clf.score(X_train, y_train)
print("Train classification accuracy: {}".format(str(np.round(train_tuned_svc_score, 2))))

# Accuracy on test data
# Accuracy improves after tuning the hyperparameters of SVC compared to using it at default parameters.
# Compare it with previously used SVC which gives training and test accuracies as .77 and .73 respectively. 
test_tuned_svc_score = tuned_svc_clf.score(X_test, y_test)
print("Test classification accuracy: {}".format(str(np.round(test_tuned_svc_score, 2))))
```

### Cross validation

```python
from sklearn.model_selection import cross_val_score
```

```python
# CV = 5 means that entire datasets is divided into 5 equal parts. 4 parts are used for training and the 5th one
# for testing. It loops over the entire dataset for 5 times whicn ensures that each sample has been used as a 
# training and test sample at different time points.
scores = cross_val_score(tuned_svc_clf, dataframe_clean_features, target_overall_survival, cv=5)
```

```python
# Accuracy on each split of cross-validation. 5 accuracy values for 5 different splits. This can be averaged to get
# one score for the entire dataset.
scores
```
